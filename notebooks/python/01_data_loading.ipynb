```python
#Import Libraries
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Project Paths
# Project directories
BASE_DIR = Path("../../")
RAW_DATA = BASE_DIR / "data/raw"
CLEAN_DATA = BASE_DIR / "data/cleaned"

print("Raw data folder:", RAW_DATA)
print("Clean data folder:", CLEAN_DATA)
```

```python
#List available raw datasets
list(RAW_DATA.glob("*.csv"))
```

```python
#list clean available datasets
list(CLEAN_DATA.glob("*.csv"))
```

```python
#Load my datasets
inflation = pd.read_csv(RAW_DATA / "inflation.csv")
gdp = pd.read_csv(RAW_DATA / "gdp_annual.csv")
monetary = pd.read_csv(RAW_DATA / "Depository Corporation Survey.csv")
trade_summary = pd.read_csv(RAW_DATA / "Foreign Trade Summary.csv")
principal_exports = pd.read_csv(RAW_DATA / "Principal Exports Volume, Value and Unit Prices (Ksh Million).csv")
value_domestic_exports = pd.read_csv(RAW_DATA / "Value of Selected Domestic Exports (Ksh Million).csv")
value_exports_african = pd.read_csv(RAW_DATA / "Value of Exports to Selected African Countries (Ksh Million).csv")
value_exports_restofworld = pd.read_csv(RAW_DATA / "Value of Exports to Selected Rest of World Countries (Ksh Million).csv")
exchange_rate = pd.read_csv(RAW_DATA / "Monthly Exchange rate (period average).csv")
```

```python
#helper function
def clean_columns(df):
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
        .str.replace("-", "_")
    )
    return df
```

```python
#apply helper function
inflation = clean_columns(inflation)
gdp = clean_columns(gdp)
monetary = clean_columns(monetary)
trade_summary = clean_columns(trade_summary)
principal_exports = clean_columns(principal_exports)
value_domestic_exports = clean_columns(value_domestic_exports)
value_exports_african = clean_columns(value_exports_african)
value_exports_restofworld = clean_columns(value_exports_restofworld)
exchange_rate = clean_columns(exchange_rate)
```

```python
#date column for time series analysis #inflation
inflation["date"] = pd.to_datetime(
    inflation["year"].astype(str) + "-" + inflation["month"].astype(str)
)

inflation = inflation.sort_values("date")

inflation.head()
```

```python
#Handling Missing values
#drop missing rows #using a dictionary
datasets = {
    "inflation": inflation,
    "gdp": gdp,
    "monetary": monetary,
    "trade_summary": trade_summary,
    "principal_exports": principal_exports,
    "value_domestic_exports": value_domestic_exports,
    "value_exports_african": value_exports_african,
    "value_exports_restofworld": value_exports_restofworld,
    "exchange_rate": exchange_rate
}

for name, df in datasets.items():
    datasets[name] = df.dropna()
```

```python
#Save clean files
gdp.to_csv(CLEAN_DATA / "gdp.csv", index=False)
monetary.to_csv(CLEAN_DATA / "monetary.csv", index=False)
trade_summary.to_csv(CLEAN_DATA / "trade_summary.csv", index=False)
inflation.to_csv(CLEAN_DATA / "inflation.csv", index=False)
principal_exports.to_csv(CLEAN_DATA / "principal_exports.csv", index=False)
value_exports_african.to_csv(CLEAN_DATA / "value_exports_african.csv", index=False)
value_exports_restofworld.to_csv(CLEAN_DATA / "value_exports_restofworld.csv", index=False)
exchange_rate.to_csv(CLEAN_DATA / "exchange_rate.csv", index=False)
```
