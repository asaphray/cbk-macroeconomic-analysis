# Kenya Macroeconomic Indicators Analysis
### Using Central Bank of Kenya (CBK) Data

## Project Overview

This project analyzes key macroeconomic indicators in Kenya using datasets published by the Central Bank of Kenya (CBK). The goal is to explore relationships between monetary policy, inflation dynamics, trade performance, and economic growth.

The analysis focuses on four pillars of the Kenyan economy:

- Monetary and Financial Statistics
- Inflation Trends
- National Accounts (GDP)
- External Trade and Balance of Payments

Using Python and R, the project performs exploratory data analysis, time series visualization, and macroeconomic interpretation of trends in the Kenyan economy.

This repository is designed to demonstratedata cleaning, exploratory analysis, economic reasoning, and reproducible research workflows.

## Research Questions

This project seeks to explore several macroeconomic questions:

### Monetary System
- How has Kenya's money supply evolved over time?
- What trends exist in credit to the private sector?

### Price Stability
- How has inflation changed across time?
- Are there seasonal or cyclical patterns in inflation?

### Economic Growth
- How has Kenya's GDP grown over the years?
- What periods show economic slowdowns or accelerations?

### External Trade
- What are Kenya's main export commodities?
- Which countries are the main destinations for Kenyan exports?

### Cross-Indicator Relationships
- Does money supply growth correlate with inflation?
- How do export trends relate to GDP growth?

## Data Sources

All datasets used in this project come from the **Central Bank of Kenya (CBK)** statistical publications.

### 1. Monetary and Financial Statistics
Depository Corporation Survey providing information on:
- Money supply aggregates
- Credit to private sector
- Banking sector balance sheets

### 2. Inflation Data
Monthly inflation statistics measuring consumer price changes.

### 3. National Accounts
Annual GDP statistics showing Kenya's economic growth.

### 4. Balance of Payments and Trade Data
Including:
- Foreign trade summary
- Principal exports (volume, value, unit price)
- Export destinations
